## Anomaly Detection Using Normalized Indicators

Raw update counts are influenced by population size. To detect true anomalies,
update volumes must be normalized relative to enrolment.


```python
df["update_burden_index"] = (
    df["total_biometric_updates"] + df["total_demographic_updates"]
) / df["total_enrolments"]

df.sort_values("update_burden_index", ascending=False).head(5)
```

### Key Insight

States with high Update Burden Index represent operational stress zones where
Aadhaar data maintenance demand is disproportionately high relative to enrolment
volume. These are strong candidates for system optimization.


```python
plt.figure(figsize=(10,5))
plt.bar(df["state"], df["update_burden_index"])
plt.xticks(rotation=90)
plt.title("Update Burden Index by State")
plt.tight_layout()
plt.show()
```
