## Aadhaar Enrolment Analysis

This section examines how Aadhaar enrolments are distributed across states.
State-wise aggregation helps identify regions with high and low Aadhaar penetration.


```python
enrol = pd.read_csv("../outputs/tables/enrolment_by_state.csv")
enrol.sort_values("total_enrolments", ascending=False).head(10)
```

### Key Insight

States with larger populations and urban concentration show higher enrolment volumes.
Lower enrolment states may indicate the need for targeted awareness or outreach
programs.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(enrol["state"], enrol["total_enrolments"])
plt.xticks(rotation=90)
plt.title("State-wise Aadhaar Enrolments")
plt.tight_layout()
plt.show()
```
